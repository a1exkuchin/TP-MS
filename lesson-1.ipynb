### Задание 1
Из колоды в 52 карты извлекаются случайным образом 4 карты. a) Найти вероятность того, что все карты – крести. б) Найти вероятность, что среди 4-х карт окажется хотя бы один туз.


### Решение:
а) всего исходов - это количество сочетаний из 52 по 4, благоприятных - количество сочетаний из 13 по 4

```python
p = comb(13, 4) / comb(52, 4)
print(f'Искомая вероятность равна {p}')
```

б) всего исходов - это количество сочетаний из 52 по 4, благоприятных - все исходы минус количество сочетаний из 48 по 4 (берем все варианты для 4-х карт кроме тех, в которых нет ни одного туза)

```python
p = 1 - comb(48, 4) / comb(52, 4)
print(f'Искомая вероятность равна {p}')
```

### Задание 2
На входной двери подъезда установлен кодовый замок, содержащий десять кнопок с цифрами от 0 до 9. Код содержит три цифры, которые нужно нажать одновременно. Какова вероятность того, что человек, не знающий код, откроет дверь с первой попытки?


### Решение:
всего исходов - это количество сочетаний из 10 по 3, благоприятных - 1

```python
p = 1 / comb(10, 3)
print(f'Искомая вероятность равна {p}')
```

### Задание 3
В ящике имеется 15 деталей, из которых 9 окрашены. Рабочий случайным образом извлекает 3 детали. Какова вероятность того, что все извлеченные детали окрашены?


### Решение:
всего исходов - это количество сочетаний из 15 по 3, благоприятных - количество сочетаний из 9 по 3

### Задание 4
В лотерее 100 билетов. Из них 2 выигрышных. Какова вероятность того, что 2 приобретенных билета окажутся выигрышными?

### Решение:
Вероятность зависит от того сколько билетов было уже продано, и сколько среди проданных было выигрышных. Будем считать, что мы покупаем первые два билета (т.е. до нас еще никто не покупал), тогда:
всего исходов - это количество сочетаний из 100 по 2, благоприятных - 1

```python
p = 1 / comb(100, 2)
print(f'Искомая вероятность равна {p}')
```
