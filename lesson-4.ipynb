### Задача 1.
Случайная непрерывная величина A имеет равномерное распределение на промежутке (200, 800].
Найдите ее среднее значение и дисперсию.
### Решение
Среднее значение находим по формуле $$M(X) = \frac{a + b}{2}.$$
Дисперсию находим по формуле $$D(X) = \frac{(b - a)^2}{12}.$$

```python
print(f'Среднее значение равно {(200+800)/2}')
print(f'Дисперсия равна {(800-200)**2/12}')
```

### Задача 2.
О случайной непрерывной равномерно распределенной величине B известно, что ее дисперсия равна 0.2.
Можно ли найти правую границу величины B и ее среднее значение зная, что левая граница равна 0.5?
Если да, найдите ее.
### Решение
По условию известно $D(X) = \frac{(b - a)^2}{12}=0.2$ и $a=0.5$. Получаем, что $b=0.5+ \sqrt {2.4}$

```python
print(f'b = {0.5+2.4**0.5}')
```

Среднее значение находим по формуле $$M(X) = \frac{a + b}{2}.$$

```python
print(f'Среднее значение равно {(0.5+0.5+2.4**0.5)/2}')
```

### Задача 4.
Рост взрослого населения города X имеет нормальное распределение.
Причем, средний рост равен 174 см, а среднее квадратичное отклонение равно 8 см.
Какова вероятность того, что случайным образом выбранный взрослый человек имеет рост:
* а). больше 182 см 
* б). больше 190 см
* в). от 166 см до 190 см
* г). от 166 см до 182 см
* д). от 158 см до 190 см
* е). не выше 150 см или не ниже 190 см
* ё). не выше 150 см или не ниже 198 см
* ж). ниже 166 см.
### Решение

```python
from math import erf

# функция нормального распределения с математическим ожиданием m и среднеквадратичным отклонением sigma
def F(x, m=174, sigma=8):
    return (1 + erf((x-m) / (sigma * 2**0.5)))/2

print(f'Вероятность того, что случайно выбранный взрослый человек имеет рост больше 182см  равна {1 - F(182)}')
print(f'Вероятность того, что случайно выбранный взрослый человек имеет рост больше 190см  равна {1 - F(190)}')

print(f'Вероятность того, что случайно выбранный взрослый человек имеет рост \
от 166см до 190см равна {F(190) - F(166)}') 

print(f'Вероятность того, что случайно выбранный взрослый человек имеет рост \
от 166см до 182см равна {F(182) - F(166)}') 

print(f'Вероятность того, что случайно выбранный взрослый человек имеет рост \
от 158см до 190см равна {F(190) - F(158)}') 

print(f'Вероятность того, что случайно выбранный взрослый человек имеет рост \
не выше 150см или не ниже 190см равна {F(150) + 1 - F(190)}')

print(f'Вероятность того, что случайно выбранный взрослый человек имеет рост \
не выше 150см или не ниже 198см равна {F(150) + 1 - F(198)}')

print(f'Вероятность того, что случайно выбранный взрослый человек имеет рост \
ниже 166см равна {F(166)}')

```

### Задача 5.
На сколько сигм (средних квадратичных отклонений) отклоняется рост человека, равный 190 см, от
математического ожидания роста в популяции, в которой M(X) = 178 см и D(X) = 25 кв.см?
### Решение
Рост человека, равный 190см, отклоняется от математического ожидания роста в популяции на $\frac{190-178}{\sqrt{25}}=\frac{12}{5}=2.4$ сигм.
