```python
from math import comb, factorial, exp
```

### Задача 1
Вероятность того, что стрелок попадет в мишень, выстрелив один раз, равна 0.8. Стрелок выстрелил 100 раз. Найдите вероятность того, что стрелок попадет в цель ровно 85 раз.
### Решение
Имеем схему Бернулли с $p=0.8,  q=0.2,  n=100$, вероятность считаем по формуле Бернулли $$P_n(m)=C_n^m \cdot p^m \cdot q^{n-m} $$

```python
p = 0.8
q = 1 - p
n = 100
m = 85
```

```python
P = comb(n, m) * p**m * q**(n-m)
print(f'Искомая вероятность равна {P}')
```

### Задача 2
Вероятность того, что лампочка перегорит в течение первого дня эксплуатации, равна 0.0004. В жилом комплексе после ремонта в один день включили 5000 новых лампочек. Какова вероятность, что ни одна из них не перегорит в первый день? Какова вероятность, что перегорят ровно две?
### Решение
Имеем схему Бернулли с $p=0.0004, n=5000$. Так как количество испытаний достаточно большое ($n=5000$) и $p<0.1$, а $\lambda=n \cdot p=2$, то формулу Бернулли можно приблизить формулой Пуассона $$P_n(m) \approx P(m)=\frac{\lambda^m}{m!}\cdot e^{-\lambda}$$

```python
p = 0.0004
q = 1 - p
n = 5000
lambd = n * p
```

```python
l = ['ек', 'ки']
for i, m in enumerate([0, 2]):
    P = lambd ** m * exp(-lambd) / factorial(m)
    print(f'Вероятность того, что в первый день выйдет из строя {m} лампоч{l[i]} равна {P}')
```

### Задача 3
Монету подбросили 144 раза. Какова вероятность, что орёл выпадет ровно 70 раз?
### Решение
Имеем схему Бернулли с $p=0.5$ (вероятность выпадания орла),  $q=0.5,  n=144$. Вероятность того, что в серии из 144 испытаний выпал 70 раз орёл, считаем по формуле Бернулли $$P_n(m)=C_n^m \cdot p^m \cdot q^{n-m} $$

```python
p = 0.5
q = 1 - p
n = 144
m = 70
```

```python
P = comb(n, m) * p**m * q**(n-m)
print(f'Искомая вероятность равна {P}')
```

### Задача 4
В первом ящике находится 10 мячей, из которых 7 - белые. Во втором ящике - 11 мячей, из которых 9 белых. Из каждого ящика вытаскивают случайным образом по два мяча. Какова вероятность того, что все мячи белые? Какова вероятность того, что ровно два мяча белые? Какова вероятность того, что хотя бы один мяч белый?
### Решение
Вероятность выбрать два белых мяча из первого ящика равна $\frac{C_7^2}{C_{10}^2}$. Вероятность выбрать два белых мяча из второго ящика равна $\frac{C_9^2}{C_{11}^2}$. Вероятность выбрать все белые мячи равна $$ \frac{C_7^2}{C_{10}^2} \cdot \frac{C_9^2}{C_{11}^2} $$


```python
P = (comb(7, 2) / comb(10, 2)) * (comb(9, 2) / comb(11, 2))
print(f'Вероятность выбрать все белые мячи равна {P}')
```

Есть следующие варианты выбора ровно двух белых мячей: 1) два белых мяча взяли из первого ящика, а два не белых - из второго, 2) два белых мяча взяли из второго ящика, а два не белых - из первого, 3) по одному белому и одному не белому мячу взяли из обоих ящиков. Количество благоприятных исходов равно $C_7^2 \cdot C_2^2 + C_9^2 \cdot C_3^2 + 7 \cdot 3 \cdot 9 \cdot 2 = 507$.
Количество всех исходов $C_{10}^2 \cdot C_{11}^2 = 2475$.
Вероятность выбрать ровно два белых мяча равна $\frac{507}{2475}$. 

```python
A = comb(7, 2) + comb(9, 2) * comb(3, 2) + 7 * 3 * 9 * 2
B = comb(10, 2) * comb(11, 2)
P = A / B
print(f'Вероятность выбрать ровно два белых мяча равна {P}')
```

Вероятность выбрать все не белые мячи равна $$p = \frac{C_3^2}{C_{10}^2} \cdot \frac{C_2^2}{C_{11}^2}. $$ Вероятность выбрать хотя бы один белый мяч $1-p$.


```python
P = 1 - (comb(3, 2) / comb(10, 2)) * (comb(2, 2) / comb(11, 2))
print(f'Вероятность выбрать хотя бы один белый мяч {P}')
```
