### Задание 1. 
Даны значения зарплат из выборки выпускников: 100, 80, 75, 77, 89, 33, 45, 25, 65, 17, 30, 24, 57, 55, 70, 75, 65, 84, 90, 150. Посчитать (желательно без использования статистических методов наподобие std, var, mean) среднее арифметическое, среднее квадратичное отклонение, смещенную и несмещенную оценки дисперсий для данной выборки.
### Решение


```python
a = array([100, 80, 75, 77, 89, 33, 45, 25, 65, 17, 30, 24, 57, 55, 70, 75, 65, 84, 90, 150])

def mean_(a):
    return sum(a) / len(a)

def var_(a, bias=False):
    m = mean_(a)
    if bias:
        var = sum((a-m)**2) / (len(a) - 1)
    else:
        var = sum((a-m)**2) / len(a)
    return var

print(f'Среднее арифметическое равно {mean_(a)}')
print(f'Выборочная (смещенная) дисперсия равна {var_(a)}')
print(f'Несмещённая (исправленная) дисперсия равна {var_(a, True)}')
print(f'Среднее квадратическое отклонение (ГОСТ Р 8.736-2011) равно {var_(a, True)**0.5}')
```

### Задание 2. 
В первом ящике находится 8 мячей, из которых 5 - белые. Во втором ящике - 12 мячей, из которых 5 белых. Из первого ящика вытаскивают случайным образом два мяча, из второго - 4. Какова вероятность того, что 3 мяча белые?
### Решение
Есть следующие варианты выбора ровно трех белых мячей: 1) два белых мяча взяли из первого ящика, один белый и три не белых - из второго, 2) один белый и один не белый мяч взяли из первого ящика, а два белых и два не белых - из второго, 3) два не белых взяли из первого ящика и один не белый и три белых мяча взяли из второго ящика. Количество благоприятных исходов равно $$C_5^2 \cdot C_5^1 \cdot C_7^3 + C_5^1 \cdot C_3^1 \cdot C_5^2 \cdot C_7^2 + C_3^2 \cdot C_5^3 \cdot C_7^1 = 5110.$$ Количество всех исходов $$C_{8}^2 \cdot C_{12}^4 = 13860.$$ Вероятность выбрать ровно три белых мяча равна $$\frac{5110}{13860} \approx 0.3687.$$

```python
n = comb(8,2) * comb(12,4)
m = comb(3,2) * comb(5,3) * comb(7,1) + comb(5,1) * comb(3,1) * comb(5,2) * comb(7,2) + comb(5,2) * comb(5,1) * comb(7,3)
p = m / n
print(f'Искомая вероятность равна {p}')
```

### Задание 3. 
На соревновании по биатлону один из трех спортсменов стреляет и попадает в мишень. Вероятность попадания для первого спортсмена равна 0.9, для второго — 0.8, для третьего — 0.6. Найти вероятность того, что выстрел произведен: a). первым спортсменом б). вторым спортсменом в). третьим спортсменом.
### Решение
Пусть событие $A$ - это попадание в мишень при одном выстреле, $H_1$ - стреляет перый спортсмен, $H_2$ - стреляет второй спортсмен,  $H_3$ - стреляет третий спортсмен. Тогда $P(H_1)=P(H_2)=P(H_3)=1/3$, по условию $P(A/H_1)=0.9, P(A/H_2)=0.8, P(A/H_3)=0.6$. $P(A)$ найдем по формуле полной вероятности $$P(A)=P(A/H_1) \cdot P(H_1) + P(A/H_2) \cdot P(H_2) + P(A/H_3) \cdot P(H_3).$$

```python
p = (0.9 + 0.8 + 0.6) / 3
print(f'P(A)={p}')
```

Вероятности $P(H_1/A), P(H_2/A), P(H_3/A)$ найдем по формуле Байеса $$P(H_1/A)=\frac{P(A/H_1) \cdot P(H_1)}{P(A)}$$

```python
l = ['первым', 'вторым', 'третьим']
pr = [0.9, 0.8, 0.6]
for i in range(3):
    print(f'Вероятность того, что выcтрел произведен {l[i]} спортсменом равна {pr[i]*(1/3)/p} ')
```

### Задание 4. 
В университет на факультеты A и B поступило равное количество студентов, а на факультет C студентов поступило столько же, сколько на A и B вместе. Вероятность того, что студент факультета A сдаст первую сессию, равна 0.8. Для студента факультета B эта вероятность равна 0.7, а для студента факультета C - 0.9. Студент сдал первую сессию. Какова вероятность, что он учится: a). на факультете A б). на факультете B в). на факультете C?
### Решение
Пусть событие $A$ - это студент сдаст первую сессию, $H_1$ - студент с факультета А, $H_2$ - студент с факультета В,  $H_3$ - студент с факультета С. Тогда $P(H_1)=P(H_2)=1/4, P(H_3)=1/2$, по условию $P(A/H_1)=0.8, P(A/H_2)=0.7, P(A/H_3)=0.9$. 
$P(A)$ найдем по формуле полной вероятности $$P(A)=P(A/H_1) \cdot P(H_1) + P(A/H_2) \cdot P(H_2) + P(A/H_3) \cdot P(H_3).$$

```python
p = (0.8 + 0.7) / 4 + 0.9 / 2
print(f'P(A)={p}')
```

Вероятности $P(H_1/A), P(H_2/A), P(H_3/A)$ найдем по формуле Байеса $$P(H_1/A)=\frac{P(A/H_1) \cdot P(H_1)}{P(A)}$$

```python
l = ['A', 'B', 'C']
pr = [0.8, 0.7, 0.9]
ps = [1/4, 1/4, 1/2]
for i in range(3):
    print(f'Вероятность того, что сдавший первую сессию студент учится на факультете {l[i]} равна {pr[i]*ps[i]/p}')
```

### Задание 5. 
Устройство состоит из трех деталей. Для первой детали вероятность выйти из строя в первый месяц равна 0.1, для второй - 0.2, для третьей - 0.25. Какова вероятность того, что в первый месяц выйдут из строя: а). все детали б). только две детали в). хотя бы одна деталь г). от одной до двух деталей?
### Решение


```python
p1 = 0.1
p2 = 0.2
p3 = 0.25
print(f'Вероятность выхода из строя всех деталей равна {p1*p2*p3}')
p4 = p1 * p2 * (1 - p3) + p1 * p3 * (1 - p2) + p3 * p2 * (1 - p1)
print(f'Вероятность выхода из строя двух деталей равна {p4}')
p5 = 1 - (1 - p1) * (1 - p2) * (1 - p3)
print(f'Вероятность выхода из строя хотя бы одной детали равна {p5}')
print(f'Вероятность выхода из строя от одной до двух деталей равна {p5-p1*p2*p3}')
```
